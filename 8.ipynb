```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np

iris = load_iris()

print("Iris Data = ", iris['target_names'])

X_train, X_test, Y_train, Y_test = train_test_split(iris['data'], iris['target'], random_state=0)

kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(X_train, Y_train)

for i in range(len(X_test)):
	prediction = kn.predict(np.array([X_test[i]]))
	print(f"\nActual = {Y_test[i]} {iris['target_names'][Y_test[i]]}")
	print(f"Predicted = {prediction} {iris['target_names'][prediction]}")

print("Test Accuracy = ", round(kn.score(X_test, Y_test), 2))
```
